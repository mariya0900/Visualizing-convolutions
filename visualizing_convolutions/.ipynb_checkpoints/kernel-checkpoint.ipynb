Here I want to try out different filters(kernels).
Note: all the digits in the filter add to 0 or 1, otherwise we would set a different weight to normalize them


```python
# Experiment with different values and see the effect
filter = [ [0, 1, 0], [1, -4, 1], [0, 1, 0]]


kernel_vert = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1],
])
kernel_horz = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1],
])

kernel_diag_fwd = np.array([
    [1, .5, 0],
    [.5, 0, -.5],
    [0, -.5, -1]
])
kernel_diag_bak = np.array([
    [0, .5, 1],
    [-.5, 0, .5],
    [-1, -.5, 0],
])

kernel_blur = np.array([
    [.05, .1, .05],
    [.1, .4, .1],
    [.05, .1, .05]
])

weight  = 1
```

Visualize the kernel

```python
import matplotlib
from matplotlib import colors
# make a figure + axes
fig, ax = plt.subplots(1, 1, tight_layout=True)
# make color map
my_cmap = matplotlib.colors.ListedColormap(['black', 'white', 'red'])

# draw the grid
for x in range(3):
    ax.axhline(x, lw=2, color='k', zorder=5)
    ax.axvline(x, lw=2, color='k', zorder=5)
# draw the boxes, display data as an image
ax.imshow(kernel_vert, interpolation='none', cmap=my_cmap, extent=[0, 3, 0, 3], zorder=0)
# turn off the axis labels
ax.axis('off')
plt.show()
```

```python
# Iterate over the image
def transform_image(image, filter):
    size_x = image.shape[0]
    size_y = image.shape[1]
    image_transformed = np.copy(image)

    for x in range(1,size_x-1):
        for y in range(1,size_y-1):
            convolution = 0.0
            convolution = convolution + (image[x-1, y-1] * filter[0][0])
            convolution = convolution + (image[x-1, y] * filter[0][1])  
            convolution = convolution + (image[x-1, y+1] * filter[0][2])     
            convolution = convolution + (image[x, y-1] * filter[1][0])    
            convolution = convolution + (image[x, y] * filter[1][1])    
            convolution = convolution + (image[x, y+1] * filter[1][2])    
            convolution = convolution + (image[x+1, y-1] * filter[2][0])    
            convolution = convolution + (image[x+1, y] * filter[2][1])    
            convolution = convolution + (image[x+1, y+1] * filter[2][2])    

            # Multiply by weight
            convolution = convolution * weight   

            # Check the boundaries of the pixel values
            if(convolution<0):
                convolution=0
            if(convolution>255):
                convolution=255
            # Load into the transformed image
            image_transformed[x, y] = convolution
    return image_transformed
```

Now let's compare original image with the transformed one, using a vertical filter

```python
view_original()
```

```python
# Plot the image
image_transformed = transform_image(ascent_image, kernel_vert)
plt.gray()
plt.grid(False)
plt.imshow(image_transformed)
plt.show()   
```

We can see that vertical lines are outlined very clearly. We can also apply max pooling to highlight those features but in a compacted image

```python
def max_pooling(image):
    size_x = image.shape[0]
    size_y = image.shape[1]
    # Assign dimensions half the size of the original image
    new_x = int(size_x/2)
    new_y = int(size_y/2)

    # Create blank image with reduced dimensions
    newImage = np.zeros((new_x, new_y))
    
    for x in range(0, size_x, 2):
      for y in range(0, size_y, 2):

        # Store all the pixel values in the (2,2) pool
        pixels = []
        pixels.append(image_transformed[x, y])
        pixels.append(image_transformed[x+1, y])
        pixels.append(image_transformed[x, y+1])
        pixels.append(image_transformed[x+1, y+1])

        # Get only the largest value and assign to the reduced image
        newImage[int(x/2),int(y/2)] = max(pixels)

    # Plot the image. Note the size of the axes -- it is now 256 pixels instead of 512
    plt.gray()
    plt.grid(False)
    plt.imshow(newImage)
    plt.show()      
```

```python
max_pooling(image_transformed)
```
